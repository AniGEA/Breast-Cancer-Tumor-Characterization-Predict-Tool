```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import joblib
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data.csv")
# df = df.drop(['id', 'Unnamed: 32'], axis=1)

keyword1 = 'worst'
columns_to_drop = [col for col in df.columns if keyword1 in col]

#

df = df.drop(columns=columns_to_drop)
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df
```

```python
final_df = pd.get_dummies(df, columns=['diagnosis'], dtype=int)
final_df
```

```python
target = final_df["diagnosis_M"]
target_names = ["1", "0"]
```

```python
features_data = df.drop("diagnosis", axis=1)
feature_names = features_data.columns
features_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_data, target, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
joblib.dump(model, 'SVM_Model.pkl')
```
